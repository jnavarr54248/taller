```python
import pandas as pd
import numpy as np

# Convertir columnas relevantes a numérico, manejando errores
columns_to_convert = [
    "TOTAL_LIABILITIES", 
    "TOTAL_ASSETS", 
    "NET_PROFIT_LOSS_FOR_THE_PERIOD", 
    "TOTAL_OPERATING_REVENUE", 
    "EBITDA_MARGIN_PERCENTAGE", 
    "INTANGIBLE_ASSETS_AND_GOODWILL"
]
for col in columns_to_convert:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convierte no numéricos a NaN

# Calcular DEBT_RATIO
df["DEBT_RATIO"] = np.where(
    (df["TOTAL_ASSETS"] - df["INTANGIBLE_ASSETS_AND_GOODWILL"]) == 0,
    0,  # Asignar 0 si el denominador es 0
    (df["TOTAL_LIABILITIES"] / (df["TOTAL_ASSETS"] - df["INTANGIBLE_ASSETS_AND_GOODWILL"])) * 100
)

# Calcular NET_MARGIN_PERCENTAGE
df["NET_MARGIN_PERCENTAGE"] = np.where(
    df["TOTAL_OPERATING_REVENUE"] == 0,
    0,  # Asignar 0 si el denominador es 0
    (df["NET_PROFIT_LOSS_FOR_THE_PERIOD"] / df["TOTAL_OPERATING_REVENUE"]) * 100
)

# Calcular EBITDA
df["EBITDA"] = np.where(
    df["EBITDA_MARGIN_PERCENTAGE"].isnull() | df["TOTAL_OPERATING_REVENUE"].isnull(),
    0,  # Asignar 0 si cualquier valor es NaN
    (df["EBITDA_MARGIN_PERCENTAGE"] / 100) * df["TOTAL_OPERATING_REVENUE"]
)

# Imprimir resultados para revisión
print(df[["DEBT_RATIO", "NET_MARGIN_PERCENTAGE", "EBITDA"]].head())
```

```python
import pandas as pd
import numpy as np

def calculate_debt_ratio_score(column):
    conditions = [
        (column < 60.00),
        (column < 65.00),
        (column < 70.00),
        (column < 72.00),
        (column < 73.00),
        (column < 74.00),
        (column < 75.00)
    ]
    scores = [8, 7, 6, 5, 4, 3, 2]
    
    return np.select(conditions, scores, default=0)

df["DEBT_RATIO_SCORE"] = calculate_debt_ratio_score(df["DEBT_RATIO"])

print(df[["DEBT_RATIO", "DEBT_RATIO_SCORE"]].head())
```

### Correlation analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import StandardScaler

# Definir la variable objetivo
df['label'] = np.where(df['OPERATIONAL_STATUS'].isin(["Liquidation", "Under Legal Investigation"]), 1, 0)

# Seleccionar solo columnas numéricas
features = df.select_dtypes(include=[np.number]).columns.tolist()
features.remove("label")

# Crear copia del DataFrame para no modificar el original
df_processed = df.copy()

# Imputación de valores nulos con mediana del sector
df_processed[features] = df_processed.groupby("INDUSTRY_CIIU")[features].transform(lambda x: x.fillna(x.median()))

# Reemplazar valores NaN e Inf restantes solo en columnas numéricas
df_processed[features] = df_processed[features].replace([np.inf, -np.inf], 10)  # Limitar valores extremos
df_processed[features] = df_processed[features].fillna(0)  # Reemplazar NaN con 0 solo en numéricas

# Normalización de los datos
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_processed[features]), columns=features)
df_scaled["label"] = df_processed["label"].values  # Agregar la variable objetivo

# Eliminar columnas constantes
constant_features = [col for col in features if df_scaled[col].std() == 0]
df_scaled.drop(columns=constant_features, inplace=True)
features = [col for col in features if col not in constant_features]

# ========  DETECCIÓN Y ELIMINACIÓN DE MULTICOLINEALIDAD Y BAJA CORRELACIÓN ========
# Calcular la matriz de correlación absoluta
correlation_matrix = df_scaled[features].corr().abs()

# Identificar pares altamente correlacionados
upper_tri = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
high_corr_features = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]

# Eliminar variables con alta correlación
df_scaled.drop(columns=high_corr_features, inplace=True)
features = [col for col in features if col not in high_corr_features]

# ========  CÁLCULO DE CORRELACIONES ========
# Pearson y Spearman con la variable objetivo
correlations = pd.DataFrame(index=features, columns=["Pearson", "Spearman", "Mutual_Info"])

for feature in features:
    if df_scaled[feature].std() > 0:  # Asegurar que la columna no es constante
        pearson_corr, _ = pearsonr(df_scaled[feature], df_scaled["label"])
        spearman_corr, _ = spearmanr(df_scaled[feature], df_scaled["label"])
        correlations.loc[feature, "Pearson"] = pearson_corr
        correlations.loc[feature, "Spearman"] = spearman_corr
    else:
        correlations.loc[feature, "Pearson"] = np.nan
        correlations.loc[feature, "Spearman"] = np.nan

# Cálculo de Información Mutua
mutual_info = mutual_info_classif(df_scaled[features], df_scaled["label"], discrete_features=False)
correlations["Mutual_Info"] = mutual_info

# Convertir valores de correlación a tipo numérico
correlations = correlations.apply(pd.to_numeric, errors='coerce')

# Ordenar por correlación absoluta más alta
correlations["Abs_Pearson"] = correlations["Pearson"].abs()
correlations["Abs_Spearman"] = correlations["Spearman"].abs()
correlations = correlations.sort_values(by="Abs_Pearson", ascending=False)

# Eliminar características con baja correlación con la variable objetivo
low_corr_features = correlations[(correlations["Abs_Pearson"] < 0.05) & (correlations["Abs_Spearman"] < 0.05)].index.tolist()
df_scaled.drop(columns=low_corr_features, inplace=True)
features = [col for col in features if col not in low_corr_features]

# ======== 4VISUALIZACIÓN ========
sns.set(style="whitegrid", font_scale=1.2)  # Mejorar el diseño de los gráficos

# Heatmap de Pearson
if not correlations["Pearson"].dropna().empty:
    plt.figure(figsize=(14, 10))
    sns.heatmap(correlations[["Pearson"]].dropna(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.6, cbar_kws={'shrink': 0.5})
    plt.title("Correlación de Pearson con la variable objetivo", fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.show()

# Heatmap de Spearman
if not correlations["Spearman"].dropna().empty:
    plt.figure(figsize=(14, 10))
    sns.heatmap(correlations[["Spearman"]].dropna(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.6, cbar_kws={'shrink': 0.5})
    plt.title("Correlación de Spearman con la variable objetivo", fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.show()

# Gráfico de barras para Información Mutua
if not correlations["Mutual_Info"].dropna().empty:
    plt.figure(figsize=(16, 10))
    sns.barplot(y=correlations.index[:30], x=correlations["Mutual_Info"].iloc[:30], palette="viridis")
    plt.xlabel("Información Mutua", fontsize=14)
    plt.ylabel("Características", fontsize=14)
    plt.title("Top 30 Características según Información Mutua", fontsize=16)
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.show()

# ======== EXPORTAR RESULTADOS ========
file_path = "correlaciones_importancia.xlsx"
correlations.to_excel(file_path, index=True)
print(f"Archivo exportado correctamente: {file_path}")

# Mostrar los resultados
print("Correlaciones y Información Mutua:")
print(correlations.head())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Función para filtrar variables altamente correlacionadas
def filter_high_correlation(df, threshold=0.85):
    """
    Elimina las variables con una correlación mayor al umbral especificado.
    :param df: DataFrame con las variables a analizar.
    :param threshold: Umbral de correlación (default=0.85).
    :return: Lista de variables seleccionadas después de filtrar correlación.
    """
    corr_matrix = df.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    # Identificar variables altamente correlacionadas
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    
    print(f"Variables eliminadas por alta correlación (> {threshold}): {to_drop}")
    return df.drop(columns=to_drop, errors='ignore')

# Cargar el DataFrame
# df = pd.read_csv("data.csv")  # Cargar dataset si es necesario
df = df.copy()
df['label'] = np.where(df['OPERATIONAL_STATUS'].isin(["Liquidation", "Under Legal Investigation"]), 1, 0)

# Variables seleccionadas
selected_variables = ["TOTAL_OPERATING_REVENUE", "OPERATING_PROFIT_EBIT", "PROFIT_BEFORE_INCOME_TAX",
                      "CASH_AND_CASH_EQUIVALENTS", "EQUITY_ATTRIBUTABLE_TO_OWNERS_OF_THE_PARENT", "TOTAL_EQUITY",
                      "ISSUED_CAPITAL", "CURRENT_LIABILITIES_TO_TOTAL_LIABILITIES_PERCENTAGE",
                      "RETURN_ON_CAPITAL_EMPLOYED_PERCENTAGE", "RETURN_ON_ASSETS_ROA_PERCENTAGE",
                      "TOTAL_ASSET_TURNOVER_X", "WORKING_CAPITAL_TURNOVER_X", "OPERATING_ROA_PERCENTAGE",
                      "CURRENT_ASSET_TURNOVER_X", "TOTAL_ASSETS", "RETURN_ON_EQUITY_ROE_PERCENTAGE",
                      "WORKING_CAPITAL", "NET_PROFIT_MARGIN_PERCENTAGE", "PROPERTY_PLANT_AND_EQUIPMENT",
                      "OPERATING_PROFIT_MARGIN_ROS_PERCENTAGE", "ADMINISTRATIVE_EXPENSES_NET_SALES_PERCENTAGE",
                      "NON_CURRENT_ASSET_TURNOVER_X", "TRADE_RECEIVABLE_TURNOVER_X", "NET_CASH", "DEBT_RATIO",
                      "TOTAL_ASSETS_TREND_PERCENTAGE", "SHAREHOLDERS_EQUITY_TREND_PERCENTAGE", "ALTMAN_Z_SCORE",
                      "GROSS_PROFIT_TREND_PERCENTAGE", "TOTAL_OPERATING_REVENUE_TREND_PERCENTAGE",
                      "NET_CASH_FLOW_FROM_USED_IN_OPERATING_ACTIVITIES", "RETAINED_EARNINGS", "CASH_RATIO_X",
                      "EBITDA", "INCOME_TAX", "OPERATING_PROFIT_TREND_PERCENTAGE", "NET_PROFIT_TREND_PERCENTAGE",
                      "DEBT_TO_EBITDA_X", "NET_CASH_FLOW_FROM_USED_IN_INVESTING_ACTIVITIES",
                      "OPERATING_CASH_FLOW_TO_DEBT_X", "SHARE_PREMIUM",
                      "NET_PROPERTY_PLANT_AND_EQUIPMENT_PPYE_TREND_PERCENTAGE", "SHORT_TERM_DEBT",
                      "INTANGIBLE_ASSETS_AND_GOODWILL", "EBITDA_TREND_PERCENTAGE", "INVENTORY_TREND_PERCENTAGE"]

# Filtrar el DataFrame con solo las variables seleccionadas
df_selected = df[selected_variables]

# Aplicar la función para filtrar variables altamente correlacionadas
df_filtered = filter_high_correlation(df_selected, threshold=0.85)

# Visualizar matriz de correlación después de filtrar
plt.figure(figsize=(12, 8))
sns.heatmap(df_filtered.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Matriz de correlación después de filtrar variables altamente correlacionadas")
plt.show()

# Lista final de variables seleccionadas
print("Variables seleccionadas:", df_filtered.columns.tolist())
```

## Clustering

```python
# Definir las variables seleccionadas para el clustering
score_columns = ['TOTAL_OPERATING_REVENUE', 'CASH_AND_CASH_EQUIVALENTS', 'EQUITY_ATTRIBUTABLE_TO_OWNERS_OF_THE_PARENT', 
                      'ISSUED_CAPITAL', 'CURRENT_LIABILITIES_TO_TOTAL_LIABILITIES_PERCENTAGE', 'RETURN_ON_CAPITAL_EMPLOYED_PERCENTAGE', 
                      'WORKING_CAPITAL', 'NET_PROFIT_MARGIN_PERCENTAGE', 'PROPERTY_PLANT_AND_EQUIPMENT', 'ADMINISTRATIVE_EXPENSES_NET_SALES_PERCENTAGE', 
                      'NON_CURRENT_ASSET_TURNOVER_X', 'TRADE_RECEIVABLE_TURNOVER_X', 'NET_CASH', 'DEBT_RATIO', 'TOTAL_ASSETS_TREND_PERCENTAGE',
                      'SHAREHOLDERS_EQUITY_TREND_PERCENTAGE', 'ALTMAN_Z_SCORE', 'GROSS_PROFIT_TREND_PERCENTAGE', 'TOTAL_OPERATING_REVENUE_TREND_PERCENTAGE',
                      'NET_CASH_FLOW_FROM_USED_IN_OPERATING_ACTIVITIES', 'RETAINED_EARNINGS', 'CASH_RATIO_X', 'EBITDA', 'INCOME_TAX', 
                      'OPERATING_PROFIT_TREND_PERCENTAGE', 'NET_PROFIT_TREND_PERCENTAGE', 'DEBT_TO_EBITDA_X', 'NET_CASH_FLOW_FROM_USED_IN_INVESTING_ACTIVITIES', 
                      'OPERATING_CASH_FLOW_TO_DEBT_X', 'SHARE_PREMIUM', 'NET_PROPERTY_PLANT_AND_EQUIPMENT_PPYE_TREND_PERCENTAGE', 'SHORT_TERM_DEBT',
                      'INTANGIBLE_ASSETS_AND_GOODWILL', 'EBITDA_TREND_PERCENTAGE', 'INVENTORY_TREND_PERCENTAGE']
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler


# ✅ Crear una copia del DataFrame original para preservar el original
df_results = df.copy()

# ✅ Filtrar solo las variables seleccionadas para clustering
df_filtered = df[score_columns].copy()

# ✅ Normalizar con MinMaxScaler para DBSCAN y K-Means
scaler = MinMaxScaler()
score_data_scaled = scaler.fit_transform(df_filtered)

# ✅ Determinar eps para DBSCAN con la curva de distancia
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(score_data_scaled)
distances, indices = neighbors_fit.kneighbors(score_data_scaled)

# 📌 Ordenar distancias para ver el punto de codo
distances = np.sort(distances[:, -1], axis=0)

plt.figure(figsize=(8, 5))
plt.plot(distances)
plt.title("Curva de Distancia para Ajustar eps en DBSCAN")
plt.xlabel("Puntos ordenados")
plt.ylabel("Distancia al vecino más cercano")
plt.show()

# ✅ Aplicar DBSCAN con eps óptimo
dbscan = DBSCAN(eps=0.5, min_samples=10)
df_results["Cluster_DBSCAN"] = dbscan.fit_predict(score_data_scaled)

print("\n📊 **Tamaño de cada Clúster con DBSCAN** 📊")
print(df_results["Cluster_DBSCAN"].value_counts())

# ✅ Aplicar PCA con el número de componentes óptimo para conservar al menos el 80% de la varianza
pca = PCA(n_components=1)  # Prueba también con 10
score_data_pca = pca.fit_transform(score_data_scaled)

explained_variance = np.sum(pca.explained_variance_ratio_) * 100
print(f"\n📌 PCA Explica el {explained_variance:.2f}% de la varianza total")

# ✅ Determinar el número óptimo de clusters en K-Means usando el método del codo
inertia = []
silhouette_scores = []
range_n_clusters = range(2, 6)

for n_clusters in range_n_clusters:
    kmeans_test = KMeans(n_clusters=n_clusters, random_state=42, n_init=10, init='k-means++')
    labels = kmeans_test.fit_predict(score_data_pca)
    inertia.append(kmeans_test.inertia_)
    silhouette_scores.append(silhouette_score(score_data_pca, labels))

# 📌 Graficar el método del codo y silhouette score
fig, ax1 = plt.subplots(figsize=(10, 5))
ax2 = ax1.twinx()

ax1.plot(range_n_clusters, inertia, 'bo-', label='Inercia (Método del Codo)')
ax2.plot(range_n_clusters, silhouette_scores, 'ro-', label='Silhouette Score')

ax1.set_xlabel('Número de Clústeres')
ax1.set_ylabel('Inercia', color='b')
ax2.set_ylabel('Silhouette Score', color='r')
plt.title("Método del Codo y Silhouette Score para Selección de Clústeres")
plt.show()

# ✅ Aplicar KMeans con el número óptimo de clusters
optimal_clusters = 4  # Ajustar según el gráfico del codo
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10, max_iter=500)
df_results["Cluster_KMeans"] = kmeans.fit_predict(score_data_pca)

print("\n📊 **Tamaño de cada Clúster con K-Means después de PCA** 📊")
print(df_results["Cluster_KMeans"].value_counts())

# ✅ Evaluar métricas de clustering para K-Means
silhouette_avg = silhouette_score(score_data_pca, df_results["Cluster_KMeans"])
calinski_harabasz = calinski_harabasz_score(score_data_pca, df_results["Cluster_KMeans"])
davies_bouldin = davies_bouldin_score(score_data_pca, df_results["Cluster_KMeans"])

# ✅ Evaluar métricas de clustering para DBSCAN (si hay más de 1 cluster)
if len(df_results["Cluster_DBSCAN"].unique()) > 1:
    silhouette_dbscan = silhouette_score(score_data_pca, df_results["Cluster_DBSCAN"])
    calinski_dbscan = calinski_harabasz_score(score_data_pca, df_results["Cluster_DBSCAN"])
    davies_dbscan = davies_bouldin_score(score_data_pca, df_results["Cluster_DBSCAN"])
else:
    silhouette_dbscan, calinski_dbscan, davies_dbscan = "N/A", "N/A", "N/A"

# ✅ Mostrar métricas de evaluación
print("\n🔍 **Evaluación del Clustering (K-Means con PCA)** 🔍")
print(f"📌 Silhouette Score: {silhouette_avg:.4f}")
print(f"📌 Calinski-Harabasz Score: {calinski_harabasz:.4f}")
print(f"📌 Davies-Bouldin Score: {davies_bouldin:.4f}")

print("\n🔍 **Evaluación del Clustering (DBSCAN)** 🔍")
print(f"📌 Silhouette Score: {silhouette_dbscan}")
print(f"📌 Calinski-Harabasz Score: {calinski_dbscan}")
print(f"📌 Davies-Bouldin Score: {davies_dbscan}")

# ✅ Resumen de variables por cluster
cluster_summary = df_results.groupby('Cluster_KMeans')[score_columns].mean()

# ✅ Mostrar resumen de los clusters con las variables
print("\n📊 **Resumen de Variables en Cada Cluster (K-Means con PCA)** 📊")
print(cluster_summary)

# ✅ Transferir los resultados del clustering de `df_results` a `df` original
df["Cluster_KMeans"] = df_results["Cluster_KMeans"]
df["Cluster_DBSCAN"] = df_results["Cluster_DBSCAN"]

# ✅ Guardar los resultados en un CSV con el `df` original actualizado
df.to_csv("resultados_clustering_con_df.csv", index=False)
print("📌 Los resultados del clustering se guardaron en 'resultados_clustering_con_df.csv'")

# ✅ **Gráfico de dispersión de los clusters con PCA en 2D**
n_components = min(2, score_data_pca.shape[1])  # Tomar el mínimo entre 2 y las dimensiones disponibles

if n_components == 2:
    pca_2d = PCA(n_components=2)
    reduced_data = pca_2d.fit_transform(score_data_pca)

    # 📌 **Visualización de la distribución de clusters**
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        x=reduced_data[:, 0], 
        y=reduced_data[:, 1], 
        hue=df_results["Cluster_KMeans"], 
        palette="Set1", 
        s=80
    )
    plt.title("📊 Distribución de Empresas por Clúster con PCA (K-Means)")
    plt.xlabel("PCA 1")
    plt.ylabel("PCA 2")
    plt.legend(title="Clúster")
    plt.show()
else:
    print("\n⚠ No se puede aplicar PCA con 2 componentes porque solo hay una dimensión en los datos transformados.")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Aplicar PCA sin límite de componentes
pca = PCA()
pca.fit(score_data_scaled)

# Obtener la varianza explicada acumulada
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Graficar la varianza explicada
plt.figure(figsize=(8,5))
plt.plot(range(1, len(explained_variance)+1), explained_variance, marker='o', linestyle='--')
plt.axhline(y=0.80, color='r', linestyle='--', label='80% Varianza Explicada')
plt.axhline(y=0.95, color='g', linestyle='--', label='95% Varianza Explicada')
plt.xlabel('Número de Componentes PCA')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Selección del Número de Componentes PCA')
plt.legend()
plt.grid()
plt.show()
```

```python
# Determinar el número óptimo de PCA con el umbral del 80% y 95%
optimal_pca_80 = np.argmax(explained_variance >= 0.80) + 1
optimal_pca_95 = np.argmax(explained_variance >= 0.95) + 1

print(f"📌 PCA recomendado para 80% de varianza: {optimal_pca_80}")
print(f"📌 PCA recomendado para 95% de varianza: {optimal_pca_95}")
```

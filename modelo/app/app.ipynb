```python
import streamlit as st
import requests

# ‚úÖ Funci√≥n para realizar la solicitud POST a la API
def realizar_solicitud_post(url, data):
    try:
        response = requests.post(url, json=data)
        if response.status_code == 200:
            return True, response.json()
        else:
            return False, response.text
    except Exception as e:
        return False, str(e)

# ‚úÖ T√≠tulo de la aplicaci√≥n
st.title("üìä Predicci√≥n de Cl√∫steres con FastAPI")

st.write("Introduce los datos de la empresa para obtener el cl√∫ster asignado.")

# ‚úÖ Formulario SIN valores predeterminados
with st.form("api_form"):
    TOTAL_OPERATING_REVENUE = st.number_input("Ingresos Operativos Totales ($)", value=None, placeholder="Ingrese el valor")
    CASH_AND_CASH_EQUIVALENTS = st.number_input("Efectivo y Equivalentes ($)", value=None, placeholder="Ingrese el valor")
    EQUITY_ATTRIBUTABLE_TO_OWNERS_OF_THE_PARENT = st.number_input("Equidad Atribuida a Propietarios ($)", value=None, placeholder="Ingrese el valor")
    ISSUED_CAPITAL = st.number_input("Capital Emitido ($)", value=None, placeholder="Ingrese el valor")
    CURRENT_LIABILITIES_TO_TOTAL_LIABILITIES_PERCENTAGE = st.number_input("Pasivos Corrientes / Total Pasivos (%)", value=None, placeholder="Ingrese el valor")
    RETURN_ON_CAPITAL_EMPLOYED_PERCENTAGE = st.number_input("Retorno sobre el Capital (%)", value=None, placeholder="Ingrese el valor")
    WORKING_CAPITAL = st.number_input("Capital de Trabajo ($)", value=None, placeholder="Ingrese el valor")
    NET_PROFIT_MARGIN_PERCENTAGE = st.number_input("Margen Neto (%)", value=None, placeholder="Ingrese el valor")
    PROPERTY_PLANT_AND_EQUIPMENT = st.number_input("Propiedad, Planta y Equipo ($)", value=None, placeholder="Ingrese el valor")
    ADMINISTRATIVE_EXPENSES_NET_SALES_PERCENTAGE = st.number_input("Gastos Administrativos / Ventas (%)", value=None, placeholder="Ingrese el valor")
    NON_CURRENT_ASSET_TURNOVER_X = st.number_input("Rotaci√≥n de Activos No Corrientes (X)", value=None, placeholder="Ingrese el valor")
    TRADE_RECEIVABLE_TURNOVER_X = st.number_input("Rotaci√≥n de Cuentas por Cobrar (X)", value=None, placeholder="Ingrese el valor")
    NET_CASH = st.number_input("Efectivo Neto ($)", value=None, placeholder="Ingrese el valor")
    DEBT_RATIO = st.number_input("Ratio de Deuda", value=None, placeholder="Ingrese el valor")
    TOTAL_ASSETS_TREND_PERCENTAGE = st.number_input("Tendencia de Activos Totales (%)", value=None, placeholder="Ingrese el valor")
    SHAREHOLDERS_EQUITY_TREND_PERCENTAGE = st.number_input("Tendencia de Equidad (%)", value=None, placeholder="Ingrese el valor")
    ALTMAN_Z_SCORE = st.number_input("Altman Z Score", value=None, placeholder="Ingrese el valor")
    GROSS_PROFIT_TREND_PERCENTAGE = st.number_input("Tendencia de Ganancia Bruta (%)", value=None, placeholder="Ingrese el valor")
    TOTAL_OPERATING_REVENUE_TREND_PERCENTAGE = st.number_input("Tendencia de Ingresos Operativos (%)", value=None, placeholder="Ingrese el valor")
    NET_CASH_FLOW_FROM_USED_IN_OPERATING_ACTIVITIES = st.number_input("Flujo de Caja Neto en Operaciones ($)", value=None, placeholder="Ingrese el valor")
    RETAINED_EARNINGS = st.number_input("Ganancias Retenidas ($)", value=None, placeholder="Ingrese el valor")
    CASH_RATIO_X = st.number_input("Ratio de Caja (X)", value=None, placeholder="Ingrese el valor")
    EBITDA = st.number_input("EBITDA ($)", value=None, placeholder="Ingrese el valor")
    INCOME_TAX = st.number_input("Impuesto sobre la Renta ($)", value=None, placeholder="Ingrese el valor")
    OPERATING_PROFIT_TREND_PERCENTAGE = st.number_input("Tendencia de Ganancia Operativa (%)", value=None, placeholder="Ingrese el valor")
    NET_PROFIT_TREND_PERCENTAGE = st.number_input("Tendencia de Ganancia Neta (%)", value=None, placeholder="Ingrese el valor")
    DEBT_TO_EBITDA_X = st.number_input("Deuda / EBITDA (X)", value=None, placeholder="Ingrese el valor")
    NET_CASH_FLOW_FROM_USED_IN_INVESTING_ACTIVITIES = st.number_input("Flujo de Caja Neto en Inversiones ($)", value=None, placeholder="Ingrese el valor")
    OPERATING_CASH_FLOW_TO_DEBT_X = st.number_input("Flujo de Caja Operativo / Deuda (X)", value=None, placeholder="Ingrese el valor")
    SHARE_PREMIUM = st.number_input("Prima sobre Acciones ($)", value=None, placeholder="Ingrese el valor")
    NET_PROPERTY_PLANT_AND_EQUIPMENT_PPYE_TREND_PERCENTAGE = st.number_input("Tendencia de Activos Fijos (%)", value=None, placeholder="Ingrese el valor")
    SHORT_TERM_DEBT = st.number_input("Deuda a Corto Plazo ($)", value=None, placeholder="Ingrese el valor")
    INTANGIBLE_ASSETS_AND_GOODWILL = st.number_input("Activos Intangibles ($)", value=None, placeholder="Ingrese el valor")
    EBITDA_TREND_PERCENTAGE = st.number_input("Tendencia de EBITDA (%)", value=None, placeholder="Ingrese el valor")
    INVENTORY_TREND_PERCENTAGE = st.number_input("Tendencia de Inventario (%)", value=None, placeholder="Ingrese el valor")

    submitted = st.form_submit_button("üì© Enviar Datos")

    if submitted:
        datos_json = {
            "TOTAL_OPERATING_REVENUE": TOTAL_OPERATING_REVENUE,
            "CASH_AND_CASH_EQUIVALENTS": CASH_AND_CASH_EQUIVALENTS,
            "EQUITY_ATTRIBUTABLE_TO_OWNERS_OF_THE_PARENT": EQUITY_ATTRIBUTABLE_TO_OWNERS_OF_THE_PARENT,
            "ISSUED_CAPITAL": ISSUED_CAPITAL,
            "CURRENT_LIABILITIES_TO_TOTAL_LIABILITIES_PERCENTAGE": CURRENT_LIABILITIES_TO_TOTAL_LIABILITIES_PERCENTAGE,
            "RETURN_ON_CAPITAL_EMPLOYED_PERCENTAGE": RETURN_ON_CAPITAL_EMPLOYED_PERCENTAGE,
            "WORKING_CAPITAL": WORKING_CAPITAL,
            "NET_PROFIT_MARGIN_PERCENTAGE": NET_PROFIT_MARGIN_PERCENTAGE,
            "PROPERTY_PLANT_AND_EQUIPMENT": PROPERTY_PLANT_AND_EQUIPMENT,
            "ADMINISTRATIVE_EXPENSES_NET_SALES_PERCENTAGE": ADMINISTRATIVE_EXPENSES_NET_SALES_PERCENTAGE,
            "NON_CURRENT_ASSET_TURNOVER_X": NON_CURRENT_ASSET_TURNOVER_X,
            "TRADE_RECEIVABLE_TURNOVER_X": TRADE_RECEIVABLE_TURNOVER_X,
            "NET_CASH": NET_CASH,
            "DEBT_RATIO": DEBT_RATIO,
            "TOTAL_ASSETS_TREND_PERCENTAGE": TOTAL_ASSETS_TREND_PERCENTAGE,
            "SHAREHOLDERS_EQUITY_TREND_PERCENTAGE": SHAREHOLDERS_EQUITY_TREND_PERCENTAGE,
            "ALTMAN_Z_SCORE": ALTMAN_Z_SCORE,
            "GROSS_PROFIT_TREND_PERCENTAGE": GROSS_PROFIT_TREND_PERCENTAGE,
            "TOTAL_OPERATING_REVENUE_TREND_PERCENTAGE": TOTAL_OPERATING_REVENUE_TREND_PERCENTAGE,
            "NET_CASH_FLOW_FROM_USED_IN_OPERATING_ACTIVITIES": NET_CASH_FLOW_FROM_USED_IN_OPERATING_ACTIVITIES,
            "RETAINED_EARNINGS": RETAINED_EARNINGS,
            "CASH_RATIO_X": CASH_RATIO_X,
            "EBITDA": EBITDA,
            "INCOME_TAX": INCOME_TAX,
            "OPERATING_PROFIT_TREND_PERCENTAGE": OPERATING_PROFIT_TREND_PERCENTAGE,
            "NET_PROFIT_TREND_PERCENTAGE": NET_PROFIT_TREND_PERCENTAGE,
            "DEBT_TO_EBITDA_X": DEBT_TO_EBITDA_X,
            "NET_CASH_FLOW_FROM_USED_IN_INVESTING_ACTIVITIES": NET_CASH_FLOW_FROM_USED_IN_INVESTING_ACTIVITIES,
            "OPERATING_CASH_FLOW_TO_DEBT_X": OPERATING_CASH_FLOW_TO_DEBT_X,
            "SHARE_PREMIUM": SHARE_PREMIUM,
            "NET_PROPERTY_PLANT_AND_EQUIPMENT_PPYE_TREND_PERCENTAGE": NET_PROPERTY_PLANT_AND_EQUIPMENT_PPYE_TREND_PERCENTAGE,
            "SHORT_TERM_DEBT": SHORT_TERM_DEBT,
            "INTANGIBLE_ASSETS_AND_GOODWILL": INTANGIBLE_ASSETS_AND_GOODWILL,
            "EBITDA_TREND_PERCENTAGE": EBITDA_TREND_PERCENTAGE,
            "INVENTORY_TREND_PERCENTAGE": INVENTORY_TREND_PERCENTAGE
        }

        url = 'http://127.0.0.1:8000/predict/'
        exito, respuesta = realizar_solicitud_post(url, datos_json)

        if exito:
            st.success(f'‚úÖ Predicci√≥n exitosa: El cl√∫ster asignado es {respuesta["cluster_predicho"]}')
        else:
            st.error(f'‚ùå Error en la predicci√≥n: {respuesta}')
```
